#2. Import Data

```python
import pandas as pd

df = pd.read_csv("loan_data.csv.csv", low_memory=False)

print(df.head())
print(df.shape)
print(df.columns.tolist())
```

#3.Identify Important Columns

```python
important_cols = [
    'loan_amnt',
    'term',
    'int_rate',
    'installment',
    'grade',
    'sub_grade',
    'emp_length',
    'home_ownership',
    'annual_inc',
    'loan_status',
    'purpose',
    'addr_state',
    'dti'
]

df = df[important_cols]
```

#4.Create Default Flag

```python
df['default_flag'] = df['loan_status'].apply(
    lambda x: 1 if x in ['Charged Off', 'Default'] else 0
)
```

#5.Check Basic Insights

```python
print(df['default_flag'].mean()*100)
```

```python
print(df.groupby('grade')['default_flag'].mean())
```

```python
print(df.groupby('purpose')['loan_amnt'].mean())
```

#6.Export Cleaned Data

```python
df.to_csv("cleaned_loan_data.csv", index=False)
```
